```python
import keras
from keras.layers import Embedding, LSTM, Dense
# Define the maximum length of the input sequences
max_length = 100
# Define the LSTM model
model = keras.Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=max_length))
model.add(LSTM(128, dropout=0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(4, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import docx

# Open the .docx file
doc = docx.Document('Project_manager_interview_Q&A.docx')

# Get the text from the document
text = ''
for para in doc.paragraphs:
    text += para.text
# Load the text data
text_data = text
# print(text_data)
# Define a custom stopword list
custom_stopwords = set(['the', 'and', 'a', 'is', 'in', 'it', 'of', 'on', 'at'])

# Tokenize the text data
tokenized_data = [word_tokenize(text) for text in text_data]

# Remove custom stopwords
filtered_data = [' '.join([word.lower() for word in tokens if word.lower() not in custom_stopwords]) for tokens in tokenized_data]

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the filtered data and transform it into numerical representations
vectorized_data = vectorizer.fit_transform(filtered_data)

# Convert the result to a dense matrix (optional)
vectorized_data = vectorized_data.toarray()
```

```python
# Train the LSTM model
model.fit(vectorized_data, epochs=10, batch_size=32, validation_split=0.2)
```
