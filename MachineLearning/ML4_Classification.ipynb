# Classification

# Evaluation Metrics in Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples = 100,n_features = 2,centers = 2,random_state = 25,cluster_std = 1.5)
```

```python
df = pd.DataFrame(X,columns = ('X1','X2'))
df['y'] = y
df[df['y'] ==0].describe()
```

```python
plt.scatter(X[:,0],X[:,1],c = y, s= 50,cmap = 'rainbow')
plt.scatter(2.5,-2,c='black' )
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
```

```python
prediction = model.predict([[2.5,-2]])
prediction
```

```python
plt.scatter(X[:,0],X[:,1],c = y, s= 50,cmap = 'rainbow')
plt.scatter(2.5,-2,c = prediction,cmap = 'rainbow' )
```

```python
Xnew = [-9,-16] + [18,25] * np.random.rand(2000,2)
```

```python
Ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:,0],X[:,1],c = y, s= 50,cmap = 'rainbow')
plt.scatter(Xnew[:,0],Xnew[:,1],c = Ynew, s= 50,cmap = 'rainbow',alpha = 0.2)
```

```python
y[:5]
```

```python
model.predict_proba(X[:5,:]).round(2)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
ypred = model.predict(X)
```

```python
print(confusion_matrix(y,ypred))
```

```python
print(accuracy_score(y,ypred))
```

```python
print(classification_report(y,ypred))
```
