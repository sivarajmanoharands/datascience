### Preparation and Shaping of the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.groupby(['Embarked'])['Fare'].mean()
```

```python
df.groupby(['Embarked'])['Fare'].count()
```

```python
df.groupby(['Embarked', 'Survived'])['Fare'].count()
```

```python
df.head(3)
```

```python
# Data Preprocessing.
# Create a Column by the name TravelAlone which would have 0 or 1 as values if a person is without any sibsp and parch
```

```python
df['TravelAlone'] = np.where((df['SibSp']+df['Parch'])>0,False,True).astype('uint8')
```

```python
df.head(3)
```

```python
df['Title'] = df.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
#'Heikkinen, Miss. Laina'.split(',')[1].split('.')[0].strip()
```

```python
df.columns
```

```python
df = df[['PassengerId',  'Pclass', 'Title','Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'TravelAlone','Survived' ]]
```

```python
df.head()
```

```python
# Dropping Unnecessary columns
```

```python
df.drop(columns= ['PassengerId','Name', 'SibSp','Parch', 'Ticket', 'Cabin'], inplace = True)
```

```python
df.head()
```

```python
df.Age.describe()
```

```python
# Missing Values
```

```python
df.isnull().sum()
```

```python
df['Age'] = df.Age.fillna(df.Age.median())
```

```python
df['Embarked'] = df.Embarked.fillna(df.Embarked.mode()[0])
```

```python
df.Embarked.mode()[0]
```

```python
df.isnull().sum()
```

```python
df.Age.describe()
```

```python
df.head()
```

```python
#catagorical variables coluns need to be encoded. There are 3 types of encoding.
#- ordinal encoding
#- labels encoding
#- one hot encoding
```

```python
newdf = pd.get_dummies(df, columns=['Title', 'Sex', 'Embarked'], drop_first=True)
newdf.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Perform normalization using MinMaxScaler
mm_scaler = MinMaxScaler()
trans_MMX = mm_scaler.fit_transform(X)
trans_MMX = pd.DataFrame(trans_MMX, columns=X.columns)
trans_MMX.describe()
sscaler = StandardScaler()
trans_SX = sscaler.fit_transform(trans_MMX)
trans_SX = pd.DataFrame(trans_SX, columns=X.columns)
trans_SX
```

```python
plt.figure(figsize=(10, 10))
for col in trans_MMX[['Age', 'Fare']]:
    plt.subplot(1, 3, 1)
    sns.kdeplot(x=col, data=X)
    plt.subplot(1,3,2)
    sns.kdeplot(x=col, data=trans_MMX)
    plt.subplot(1,3,3)
    sns.kdeplot(x=col, data=trans_SX)
    plt.show()
```
